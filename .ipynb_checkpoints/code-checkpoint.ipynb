```python
import yfinance as yf
import pandas as pd
import numpy as np
import sklearn.preprocessing
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt

# ensure the code run consiseten
import numpy as np
import tensorflow as tf

np.random.seed(42)
tf.random.set_seed(42)

# fetch data with previous 3 years
def fetch_stock_data(ticker_symbol):
    end_date = pd.to_datetime("today").strftime('%Y-%m-%d')
    start_date = (pd.to_datetime(end_date) - pd.DateOffset(years=3)).strftime('%Y-%m-%d')
    return yf.download(ticker_symbol, start=start_date, end=end_date)

# data normalization
def normalize_data(df):
    min_max_scaler = sklearn.preprocessing.MinMaxScaler()
    df['Open'] = min_max_scaler.fit_transform(df.Open.values.reshape(-1,1))
    df['High'] = min_max_scaler.fit_transform(df.High.values.reshape(-1,1))
    df['Low'] = min_max_scaler.fit_transform(df.Low.values.reshape(-1,1))
    close_scaler = sklearn.preprocessing.MinMaxScaler()
    df['Close'] = close_scaler.fit_transform(df['Close'].values.reshape(-1,1))
    return df, close_scaler

# split data into training and validation sets
def load_data(stock, seq_len):
    data_raw = stock.values
    data = []
    for index in range(len(data_raw) - seq_len):
        data.append(data_raw[index: index + seq_len])
    data = np.array(data)
    valid_set_size = int(np.round(0.1 * data.shape[0]))
    train_set_size = data.shape[0] - valid_set_size
    x_train = data[:train_set_size,:-1,:]
    y_train = data[:train_set_size,-1,:]
    x_valid = data[train_set_size:,:-1,:]
    y_valid = data[train_set_size:,-1,:]
    return [x_train, y_train, x_valid, y_valid]

# adjust LSTM model
def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(150, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.5))
    model.add(LSTM(150, return_sequences=True))
    model.add(Dropout(0.5))
    model.add(LSTM(75, return_sequences=False))
    model.add(Dense(4))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# train
def train_and_predict_stock_prices(stock_data_norm):
    x_train, y_train, x_valid, y_valid = load_data(stock_data_norm, seq_len=40)
    model = build_model((x_train.shape[1], x_train.shape[2]))
    model.fit(x_train, y_train, epochs=100, batch_size=64, validation_data=(x_valid, y_valid), shuffle=False, verbose=0)
    predictions = []
    current_batch = x_valid[-1:].copy()
    for i in range(10):
        current_pred = model.predict(current_batch)[0]
        predictions.append(current_pred)
        current_batch = np.append(current_batch[:,1:,:], [[current_pred]], axis=1)
    return predictions

# plotting
def plot_past_and_predicted(stock_past, actual_predictions, ticker):
    plt.figure(figsize=(15, 6))
    plt.plot(stock_past.index, stock_past['Close'], label='Past Prices', color='blue')
    predicted_dates = pd.date_range(stock_past.index[-1], periods=11, closed='right')
    plt.plot(predicted_dates, actual_predictions, label='Predicted Prices', color='red', linestyle='dashed')
    plt.title(f"{ticker} Stock Price: Past and Predicted")
    plt.ylabel('Price')
    plt.xlabel('Date')
    plt.legend()
    plt.grid(True)
    plt.show()

# fetch particular data and traning and predicting
msft_data = fetch_stock_data("MSFT")
tsla_data = fetch_stock_data("TSLA")

msft_data_norm, msft_close_scaler = normalize_data(msft_data[['Open', 'High', 'Low', 'Close']])
tsla_data_norm, tsla_close_scaler = normalize_data(tsla_data[['Open', 'High', 'Low', 'Close']])

msft_predictions = train_and_predict_stock_prices(msft_data_norm)
tsla_predictions = train_and_predict_stock_prices(tsla_data_norm)

actual_msft_predictions = msft_close_scaler.inverse_transform(np.array(msft_predictions)[:, 3].reshape(-1, 1)).flatten()
actual_tsla_predictions = tsla_close_scaler.inverse_transform(np.array(tsla_predictions)[:, 3].reshape(-1, 1)).flatten()

# plot outcome
plot_past_and_predicted(msft_data[-20:], actual_msft_predictions, "MSFT")
plot_past_and_predicted(tsla_data[-20:], actual_tsla_predictions, "TSLA")
```
