```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sksurv.datasets import load_whas500
from sksurv.linear_model import CoxPHSurvivalAnalysis
from sksurv.preprocessing import OneHotEncoder
```

```python
data_x, data_y = load_whas500()
```

```python
data_x
```

```python
data_y[0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=50, random_state=2020)
```

```python
pipeline = make_pipeline(OneHotEncoder(), CoxPHSurvivalAnalysis())
```

```python
pipeline.fit(X_train, y_train)
```

```python
c_index = pipeline.score(X_test, y_test)
```

```python
c_index
```

```python
surv_fns = pipeline.predict_survival_function(X_test)
```
